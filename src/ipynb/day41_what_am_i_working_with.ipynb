```python
# Load libs
import psycopg2
import pandas.io.sql as pdsql

# Specify our database
dbname="kick"

# Connect to database
conn = psycopg2.connect(dbname=dbname)
```

```python
# Create table
name_of_table = "info"

cmd = "CREATE TABLE %s (" % name_of_table + "\n"
for n,t in zip(df.columns, df.dtypes):
    if n == "id":
        data_type = "INTEGER PRIMARY KEY"
    elif t == object:
        data_type = "VARCHAR"
    elif t == float:
        data_type = "FLOAT"
    elif t == bool:
        data_type = "BOOLEAN"
    elif t == int:
        data_type = "INTEGER"
        
    cmd += "    {} {},".format(n, data_type) + "\n"
    
cmd = cmd.strip(",\n") + "\n);"
print(cmd)
```

```python
cur = conn.cursor()
cur.execute(cmd)
```

```python
# Make the changes to the database persistent
conn.commit()

# Close communication with the database
cur.close()
conn.close()
```
